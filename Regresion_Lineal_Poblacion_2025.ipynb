```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Cargar los datos desde tu base de datos
df = pd.read_csv('https://raw.githubusercontent.com/FelipeHerreraPinzon/Data_Sharks_Project/main/world_population.csv')
#df
```

```python
# Lista de columnas a sumar
years_columns = ['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population', '2020 Population', '2022 Population']

# Sumar los valores de cada columna
suma_columnas = df[years_columns].sum()

# Crear un nuevo DataFrame con los resultados de la suma
df_suma_poblacion = pd.DataFrame(suma_columnas, columns=['poblacion_total'])

# Transponer el DataFrame para que sea una fila en lugar de una columna
df_suma_poblacion = df_suma_poblacion.T
df_suma_poblacion

# Convertir los valores del DataFrame en una lista sin anidamiento
lista_total_poblacion = df_suma_poblacion.values.flatten().tolist()
lista_total_poblacion


years = ["1970", "1980", "1990", "2000", "2010", "2015", "2020", "2022"]

fig, ax = plt.subplots()
ax.scatter(years, lista_total_poblacion, label="Población Mundial", color="red")


ax.legend()
# Mueve la leyenda fuera de la gráfica
ax.legend(bbox_to_anchor=(1, 1), loc='upper left')


plt.show()
```

```python


# Convertir la lista de años y la lista de población a arrays de NumPy
years_array = np.array([int(year) for year in years]).reshape(-1, 1)
poblacion_array = np.array(lista_total_poblacion)

# Inicializar y entrenar el modelo de regresión lineal
model = LinearRegression().fit(years_array, poblacion_array)

# Obtener la pendiente y la intersección del modelo
pendiente = model.coef_[0]
interseccion = model.intercept_

# Imprimir la ecuación de la recta de regresión
print("Ecuación de la recta de regresión:")
print(f"Población = {pendiente} * Año + {interseccion}")

# Predecir la población para el año 2025 (por ejemplo)
poblacion_2025 = model.predict([[2025]])
print("Predicción de población para el año 2025:", poblacion_2025[0])

# Graficar los datos y la línea de regresión
import matplotlib.pyplot as plt

plt.scatter(years, lista_total_poblacion, label="Población Mundial", color="red")
plt.plot(years, model.predict(years_array), label="Regresión Lineal", color="blue")
plt.xlabel("Año")
plt.ylabel("Población Mundial")
plt.title("Regresión Lineal de Población Mundial")
plt.legend()
plt.grid(True)
plt.show()
```
